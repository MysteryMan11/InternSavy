```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set() 
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('../segmentation data.csv')
df.head(5)
```

```python
df.info()
```

```python
df.describe().T
```

```python
plt.figure(figsize=(12,9))
mask = np.triu(np.ones_like(df.corr(), dtype=bool))
sns.heatmap(df.corr(), vmin=-1, vmax=1, mask=mask, annot=True, cmap='RdBu')
plt.title('Correlation Heatmap',fontsize=14)
plt.yticks(rotation =0)
plt.show()
```

```python
plt.figure(figsize = (12, 9))
plt.scatter(df.iloc[:, 2], df.iloc[:, 4])
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Visualization of raw data')
```

```python
scaler = StandardScaler()
df_std = scaler.fit_transform(df)
```

```python
df_std = pd.DataFrame(data = df_std,columns = df.columns)
```

```python
pca = PCA()
pca.fit(df_std)
```

```python
pca.explained_variance_ratio_
```

```python
pca.transform(df_std)
scores_pca = pca.transform(df_std)
```

```python
wcss = []
for i in range(1,11):
    kmeans_pca = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans_pca.fit(scores_pca)
    wcss.append(kmeans_pca.inertia_)

```

```python
plt.figure(figsize = (10,8))
plt.plot(range(1, 11), wcss, marker = 'o', linestyle = '-.',color='red')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('K-means Clustering')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 4, init = 'k-means++', random_state = 42)
```

```python
kmeans.fit(df_std)
```

```python
df_segm_kmeans= df_std.copy()
df_std['Segment K-means'] = kmeans.labels_
```

```python
df_segm_analysis = df_std.groupby(['Segment K-means']).mean()
df_segm_analysis.T
```
